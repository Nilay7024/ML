```python
df.drop(['season'],axis =1,inplace =True)
```

```python
df.drop(['casual','registered'],axis=1,inplace =True)
```

```python
#giving target variable and features
```

```python
X = df.drop(['cnt'],axis =1)
Y = df['cnt']
```

```python
#splitting train and test data
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test = train_test_split(X,Y,test_size = 0.25,random_state = 42)
```

```python
#standard scaling of values
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
x_test= scale.transform(x_test)
```

```python
#model creation
```

```python
from sklearn.tree import DecisionTreeRegressor
rt = DecisionTreeRegressor(min_samples_split = 25,random_state = 102)
rt_model = rt.fit(X_train,Y_train)
rt_pred = rt_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,rt_pred)
```
