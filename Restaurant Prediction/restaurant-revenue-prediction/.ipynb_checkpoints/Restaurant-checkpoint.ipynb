```python
pre_y = {
    2013:20,
    2012:19,
    2010:18,
    2009:17,
    2011:16,
    2007:15,
    2008:14,
    2006:13,
    2000:12,
    1998:11,
    1997:10,
    1999:9,
    2004:8,
    2005:7,
    1996:6,
    1995:5,
    2003:4,
    2001:3,
    2002:2,
    2014:1
}
def change(col):
    return pre_y.get(col)

rest["Y"] = rest["Year"].apply(change)
rest.head()
```

```python
pre_m = {
    12:12,
    11:11,
    8:10,
    6:9,
    10:8,
    9:7,
    3:6,
    7:5,
    4:4,
    5:3,
    2:2,
    1:1
}

def change(col):
    return pre_m.get(col)

rest["M"] = rest["Month"].apply(change)
rest.head()
```

```python
rest.drop(columns=["Month","Year"],inplace=True)
```

```python
R_Type = pd.get_dummies(rest["Type"],drop_first=True)
```

```python
R_City = pd.get_dummies(rest["City"], drop_first=True)
```

```python
rest=pd.concat([rest,R_Type,R_City],axis=1)
rest.head()
```

```python
rest.drop(columns=["Id", "Type", "City"], inplace=True)
rest.head()
```

```python
rest.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X = rest.drop(columns=["revenue"])
Y = rest["revenue"]

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42,test_size=0.3)

restaurant = RandomForestRegressor(max_depth=4, random_state=42)
restaurant.fit(X_train, y_train)

```

```python
X_train = X_train.values.reshape(-1,1)
y_train = y_train.values.reshape(-1,1)
```

```python
y_pred = restaurant.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```

```python
restaurant.score(X_test, y_test)
```
