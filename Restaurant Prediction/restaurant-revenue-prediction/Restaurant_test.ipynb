## In testing dataset we have 4 category whereas in training dataset we have three and among them MB is not present, So here we replace the MB category with the FC category as it is more frequent than the others 

```python
def change(x):
    if(x=="MB"):
        return "FC"
    else:
        return x
    
rest["Type"] = rest["Type"].apply(change)
rest.head()
```
