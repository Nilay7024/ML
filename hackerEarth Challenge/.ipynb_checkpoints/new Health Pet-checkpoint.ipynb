```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
x = train.drop(columns=["breed_category","pet_category"])
y = train["breed_category"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
GBC = GradientBoostingClassifier(random_state=42,max_depth=5)
GBC.fit(x_train,y_train)
```

```python
GBC.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
X = train.drop(columns=["breed_category","pet_category"])
Y = train["pet_category"]

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
train_gbc = GradientBoostingClassifier(random_state=42)
train_gbc.fit(X_train,Y_train)
```

```python
train_gbc.score(X_test,Y_test)
```
